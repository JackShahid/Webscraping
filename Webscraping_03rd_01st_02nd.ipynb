# Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
 # creating a empty list
Brand_Name = []
Product_Name = []
Rating = []
No_of_Rating = []
Price = []
Return = []
Delivery = []
Availability = []
Details = []
Product_url = []
```

```python
brand1 = driver.find_elements_by_xpath("//h2[@class = 'a-size-mini a-spacing-none a-color-base s-line-clamp-3']")
for i in brand1:
    Brand_Name.append(i.text)
Brand_Name
```

```python
product1 = driver.find_elements_by_xpath("//a[@class = 'a-link-normal a-text-normal']")
for i in product1:
    Product_Name.append(i.text)
Product_Name
```

```python
urls= driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
UR=[]
for i in urls:
    UR.append(i.get_attribute('href'))
for url in UR:
    driver.get(url)
    try:
        deliveries= driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']")
        Delivery.append(deliveries.text)
        
        rate= driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        rate.click()
        
        rating= driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Rating.append(rating.text)
        
        rates= driver.find_element_by_xpath("//span[@class='a-size-base a-color-secondary']")
        No_of_Rating.append(rates.text)
        
        price = driver.find_element_by_xpath("//span[@class = 'a-price-whole']")
        Price.append(price.text)
        
        return1 = driver.find_element_by_xpath("//a[@class ='a-size-small a-link-normal a-text-normal']")
        Return.append(return1.text)
        
        available = driver.find_element_by_xpath("//div[@class = 'a-section a-spacing-base }']")
        Availability.append(available.text)
        
        details = driver.find_element_by_xpath("//div[@class = 'a-section a-spacing-medium a-spacing-top-small']")
        Details.append(details.text)
        
    except NoSuchElementException   as e:
        
        Rating.append("--")      #appending the -- if no rating is there
        No_of_Rating.append("--")
        Delivery.append("--")
        Price.append("--")
        Return.append("--")
        Availability.append("--")
        Details.append("--")
```
