# 8. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
import pandas as pd
import selenium 
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
url= ("https://www.forbes.com/")
driver.get(url)
```

```python
button = driver.find_element_by_xpath("//button[@class = 'icon--hamburger']")
button.click()
```

```python
rank = driver.find_elements_by_xpath("//div[@class = 'rank']")
rank
```

```python
Rank = []
for i in rank:
    Rank.append(i.text)
Rank
```

```python
len(Rank)
```

```python
name = driver.find_elements_by_xpath("//div[@class = 'personName']")
Name = []
for i in name:
    Name.append(i.text)
Name
```

```python
len(Name)
```

```python
worth = driver.find_elements_by_xpath("//div[@class = 'netWorth']")
Net_worth = []
for i in worth:
    Net_worth.append(i.text)
Net_worth
```

```python
len(Net_worth)
```

```python
country = driver.find_elements_by_xpath("//div[@class = 'countryOfCitizenship']")
Citizenship = []
for i in country:
    Citizenship.append(i.text)
Citizenship
```

```python
len(Citizenship)
```

```python
source = driver.find_elements_by_xpath("//div[@class = 'source-column']")
Source = []
for i in source:
    Source.append(i.text)
Source
```

```python
len(Source)
```

```python
richest = pd.DataFrame({})
richest['Rank'] = Rank
richest['Name'] = Name
richest['Net_worth'] = Net_worth
richest['Citizenship'] = Citizenship
richest['Source'] = Source
richest
```
