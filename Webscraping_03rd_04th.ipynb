# 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.


```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
```

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
url = ("https://www.flipkart.com/")
driver.get(url)
```

```python
search_bar= driver.find_element_by_class_name('_3704LK')
search_bar.clear()
search_bar.send_keys('oneplus Nord, pixel 4A')
```

```python
button= driver.find_element_by_class_name('_34RNph')
button.click()
```

```python
brand=[]
brands= driver.find_elements_by_class_name('_4rR01T')
for i in brands:
    brand.append(i.text)
brand
```

```python
ratings=[]
rating= driver.find_elements_by_class_name('_1lRcqv')#scraping description by class name = '_2mylT6'
for i in rating:
    ratings.append(i.text)
ratings
```

```python
RAMs=[]
RAM = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[1]")# scraping the discount from the xpath
for i in RAM:
    RAMs.append(i.text)
RAMs
```

```python
inches=[]
inch= driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[2]")# scraping the discount from the xpath
for i in inch:
    inches.append(i.text)
inches
```

```python
camera=[]
cameras= driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[3]")# scraping the discount from the xpath
for i in cameras:
    camera.append(i.text)
camera
```

```python
battery=[]
batteries= driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[4]")# scraping the discount from the xpath
for i in batteries:
    battery.append(i.text)
battery
```

```python
price=[]
prices= driver.find_elements_by_xpath("//div[@class='_3utEwz']")# scraping the price from the xpath
for i in prices:
    price.append(i.text)
price
```

```python
color =[]
colors = driver.find_elements_by_xpath("//div[@class='_4rR01T']")# scraping the price from the xpath
for i in colors:
    color.append(i.text)
color
```

```python
mobile = pd.DataFrame({})

mobile['brand']= brand[:20]
mobile['rating'] = ratings[:20]
mobile['Display'] = inches[:20]
mobile['Camera']= camera[:20]
mobile['price'] = price[:20]
mobile['RAM'] = RAMs[:20]
mobile['battery'] = battery[:20]
mobile['Color'] = color[:20]
mobile
```
